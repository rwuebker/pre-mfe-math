```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
```

```python
def build_process(n=4):
    fs = {}
    fs[1] = {0:0, 0.5: 0.5, 1.0:1.0}
    plt.plot(list(fs[1].keys()), list(fs[1].values()))
    
    for j in range(2, n + 1):
        k = 0
        J = j - 1
        domain_len = 2**j + 1
        domain = np.linspace(0, 1, domain_len)
        f = {}
        for ix, t in enumerate(domain):
            if t in fs[j-1]:
                f[t] = fs[j-1][t]
            else:
                k += 1
                prev = (fs[j-1][domain[ix-1]] + fs[j-1][domain[ix+1]])/2
                new_val = prev + (-1)**(k + J + 1) * 2**((-J/2) - 1)
                f[t] = new_val
        
        fs[j] = f
        plt.plot(list(fs[j].keys()), list(fs[j].values()))
    return fs
        
fs = build_process(8)
```
